# Machine Learning

```python
from sklearn.metrics import classification_report, accuracy_score
```

## Data Splitting

```python
from sklearn.model_selection import train_test_split

X = data.drop(['quality'], axis=1)
y = data['quality']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

## Standard Scaler

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)
```

## Models

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(solver='lbfgs', multi_class='auto')

logreg.fit(X_train, y_train)
pred_logreg = logreg.predict(X_test)
```

```python
logreg_acc_score = accuracy_score(y_test, pred_logreg)
logreg_class_report = classification_report(y_test, pred_logreg)

print('Accuracy Score:', logreg_acc_score)
print(classification_report(y_test, pred_logreg))
```

### K Nearest Neighbor

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=20)

knn.fit(X_train, y_train)
pred_knn = knn.predict(X_test)
```

```python
knn_acc_score = accuracy_score(y_test, pred_knn)
knn_class_report = classification_report(y_test, pred_knn)

print('Accuracy Score:', knn_acc_score)
print(classification_report(y_test, pred_knn))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier()

dtree.fit(X_train, y_train)
pred_dtree = dtree.predict(X_test)
```

```python
dtree_acc_score = accuracy_score(y_test, pred_dtree)
dtree_class_report = classification_report(y_test, pred_dtree)

print('Accuracy Score:', dtree_acc_score)
print(classification_report(y_test, pred_dtree))
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=100)

rfc.fit(X_train, y_train)
pred_rfc = rfc.predict(X_test)
```

```python
rfc_acc_score = accuracy_score(y_test, pred_rfc)
rfc_class_report = classification_report(y_test, pred_rfc)

print('Accuracy Score:', rfc_acc_score)
print(classification_report(y_test, pred_rfc))
```

#### Feature Importance

```python
feature_importance = pd.DataFrame({'Features': X_train.columns, 
                                  'Importance': rfc.feature_importances_})
```

```python
plt.figure(figsize = (18,7))
sns.barplot(x="Importance", y="Features", data=feature_importance.sort_values(by=['Importance']))
```

### Model Summary

```python
model_summary = pd.DataFrame({'models': ["Logistic Regression",
                                      "KNeighborsClassifier",
                                      "DecisionTreeClassifier",
                                      "RandomForestClassifier"],
                           'accuracies': [logreg_acc_score,
                                          knn_acc_score,
                                          dtree_acc_score,
                                          rfc_acc_score,]})
```

```python
model_summary.sort_values('accuracies' ,ascending=False)
```

```python
plt.figure(figsize = (18,7))
sns.barplot(x="accuracies", y="models", data=model_summary.sort_values(by=['accuracies']))
```

### K fols cross-validation

```python
from sklearn.model_selection import cross_val_score

scores_rfc = cross_val_score(rfc, X_train, y_train, cv=10, scoring='accuracy')

print(scores_rfc.mean())
```

### Grid Search 

```python
grid_param = {  
    'n_estimators': [100, 300, 500, 800, 1000],
    'criterion': ['gini', 'entropy'],
    'bootstrap': [True, False]
}
```

```python
from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(estimator=rfc,  
                     param_grid=grid_param,
                     scoring='accuracy',
                     cv=5,
                     n_jobs=-1)

grid_search.fit(X_train, y_train) 
```

```python
best_parameters = grid_search.best_params_  
print(best_parameters)  
```

```python
best_result = grid_search.best_score_  
print(best_result)  
```
